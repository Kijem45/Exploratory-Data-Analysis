```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
office=pd.read_csv("dm_office_sales.csv")
office
```

```python
sns.scatterplot(data=office, x="sales", y="salary")   #This shows relationship between two variables.
#the output shows that as sales is growing, salary is also growing.
```

```python
sns.scatterplot(data=office, x="sales", y="salary", hue="level of education")
```

```python
# This shows a normal distribution in salary
sns.displot(data=office, x="salary")
plt.show() 
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=office, x="division")
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=office, x="division", hue="training level", palette="viridis")
```
